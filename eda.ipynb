## Codigo para verificar que  hay algun tipo de correlacion entre las variables - (Las tablas finales se muestran en la presentacion) - El output de este codigo

```python
from scipy.stats import chi2_contingency
import numpy as np
def prueba_chi_cuadrado_1(df, columna1, columna2):
  # Crear tabla SIN marginales para el cálculo
    tabla_sin_marginales = pd.crosstab(df[columna1], df[columna2])
    #print(tabla_sin_marginales)

    # Tabla CON marginales para mostrar
    tabla_con_marginales = pd.crosstab(df[columna1], df[columna2],
                                     margins=True, margins_name='Total')

    # Realizar prueba Chi-Cuadrado
    chi2, p_value, dof, expected = chi2_contingency(tabla_sin_marginales)

    # Calcular V de Cramer (usando tabla sin marginales)
    n = len(df)
    v_cramer = np.sqrt(chi2 / (n * (min(tabla_sin_marginales.shape) - 1)))
    if p_value < 0.05:
      print("Tabla de Contingencia:")
      print(tabla_sin_marginales)
      print("\n")
      print(f"Resultado SIGNIFICATIVO (p = {p_value:.4f})")
      print(f"V de Cramer: {v_cramer:.6f}")
      print(f"La relación entre condición {columna1} y {columna2}")
      print("\n" + "="*50)
      return tabla_sin_marginales, True
    else:
      return False

def grafico_barras_agrupadas(tabla):
  fig, ax = plt.subplots(figsize=(12, 7))
  tabla.plot(kind='bar', ax=ax, linewidth=0.5)
  plt.title('Gráfico de Barras Agrupadas')
  plt.xlabel('')
  plt.ylabel('Frecuencia')
  plt.legend(title='Región')
  plt.xticks(rotation=45)
  ax.grid(axis='y', alpha=0.3, linestyle='-', linewidth=0.8)
  #ax.grid(axis='y', which='minor', alpha=0.2, linestyle=':', linewidth=0.5)
  #ax.yaxis.set_minor_locator(plt.MultipleLocator(5))
  for container in ax.containers:
        ax.bar_label(container, fmt='%d', fontsize=9, padding=2,
                    label_type='edge', color='darkblue', fontweight='bold')
  plt.tight_layout()
  plt.show()

def grafico_barras_porcentaje(tabla_freq):
  tabla_porc = tabla_freq.div(tabla_freq.sum(axis=1), axis=0) * 100
  fig, ax = plt.subplots(figsize=(12, 7))
  tabla_porc.plot(kind='bar', stacked = True, ax=ax, linewidth=0.5)
  plt.title('Gráfico de Barras Apiladas 100%')
  plt.xlabel('')
  plt.ylabel('Porcentaje (%)')
  plt.legend(title='Región', bbox_to_anchor=(1.05, 1), loc='upper left')
  plt.xticks(rotation=45)
  ax.grid(axis='y', alpha=0.3, linestyle='-', linewidth=0.8)
  for i, (idx, row) in enumerate(tabla_porc.iterrows()):
        # Calcular la posición acumulada para cada segmento
        acumulado = 0
        for j, (col, valor) in enumerate(row.items()):
            if valor > 5:  # Solo mostrar si el porcentaje es mayor a 5% (para evitar texto amontonado)
                # Posición vertical centrada en el segmento
                y_pos = acumulado + (valor / 2)

                # Color del texto (blanco para segmentos oscuros, negro para claros)
                color_texto = 'white' if valor > 50 else 'black'

                ax.text(i, y_pos, f'{valor:.1f}%',
                       ha='center', va='center',
                       fontsize=9, fontweight='bold',
                       color=color_texto)

            acumulado += valor
  plt.tight_layout()
  plt.show()

relaciones_significativas = 0
total_comparaciones = 0

columnas_filtradas_categoricas = df_categorical.columns.tolist()

for i in range(len(columnas_filtradas_categoricas)):
    for j in range(i + 1, len(columnas_filtradas_categoricas)): ## Asi evitamos realizar duplicados al recorrer la lista de esta manera
        columna1 = columnas_filtradas_categoricas[i]
        columna2 = columnas_filtradas_categoricas[j]
        if columna1 != columna2:
            total_comparaciones += 1
            if prueba_chi_cuadrado_1(df_categorical, columna1, columna2):
                resultado = prueba_chi_cuadrado_1(df_categorical, columna1, columna2)
                tabla_contingencia = resultado[0]
                grafico_barras_agrupadas(tabla_contingencia)
                grafico_barras_porcentaje(tabla_contingencia)

                relaciones_significativas += 1
if relaciones_significativas == 0:
  print("Total comparaciones: ", total_comparaciones)
  print("No hay relaciones significativas entre las variables")
        
```

```python
# Prueba de poder estadístico
from statsmodels.stats.power import GofChisquarePower

def analizar_poder_muestral(tabla_observada, alpha=0.05):
    """
    Analiza si el tamaño muestral es suficiente para detectar efectos
    """
    n = tabla_observada.sum().sum()  # Tamaño total de la muestra
    dof = (tabla_observada.shape[0] - 1) * (tabla_observada.shape[1] - 1)

    # Calcular poder para diferentes tamaños de efecto
    power_analysis = GofChisquarePower()

    print(" ANÁLISIS DE PODER MUESTRAL")
    print("="*40)
    print(f"Tamaño muestral (n): {n}")
    print(f"Grados de libertad: {dof}")
    print(f"Nivel de significancia (α): {alpha}")

    # Poder para diferentes tamaños de efecto (w)
    efectos = [0.1, 0.3, 0.5]  # pequeño, mediano, grande- Mide el efecto de la asociacion entre las variables, asi asociacion debil, fuerte, grande(Fumador vs Enfermedad Pulmonar)
    for w in efectos:
        poder = power_analysis.solve_power(effect_size=w, nobs=n, alpha=alpha, n_bins=dof+1)
        print(f"Poder para efecto {w}: {poder:.3f}")

    # ¿Qué tamaño muestral se necesitaría?
    print(f"\n TAMAÑO MUESTRAL RECOMENDADO:")
    for w in efectos:
        n_necesario = power_analysis.solve_power(effect_size=w, power=0.8, alpha=alpha, n_bins=dof+1)
        print(f"Para detectar efecto {w}: {int(n_necesario)} observaciones")

#for columna1 in columnas_filtradas_categoricas:
#    for columna2 in columnas_filtradas_categoricas:
#        if columna1 != columna2:
#          print(f"VARIABLES :{columna1} vs {columna2}")
#          tabla = pd.crosstab(df_categorical[columna1], df_categorical[columna2])
#          analizar_poder_muestral(tabla)
```
